```python
import pandas as pd
import os
import glob
```

```python

path = os.getcwd()
csv_files = glob.glob(os.path.join(path + '/ejecucion/', "*.csv"))
#csv_files = [csv_files[4]]
csv_files
```

```python
dfejecucion = pd.DataFrame()
for f in csv_files:
    df = pd.read_csv(f, sep=';', thousands='.')
    dfejecucion = pd.concat([dfejecucion, df], axis=0)

dfejecucion.reset_index(drop=True, inplace=True)
dfejecucion.columns = ['Ano', 'Mes', 'Moneda', 'Partida', 'Nombre Partida', 'Capitulo', 'Nombre Capitulo', 'Programa', 'Nombre Programa', 'Subtitulo', 'Nombre Subtitulo', 'Item', 'Nombre Item','Asignacion', 'Nombre Asignacion', 'Sub-Asignación', 'Nombre Sub-Asignación', 'Ejecución de AGOSTO', 'Ejecución a AGOSTO']
```

```python
dfejecucion['Item'].fillna(0, inplace=True)
dfejecucion['Asignacion'].fillna(0, inplace=True)
dfejecucion['Sub-Asignación'].fillna(0, inplace=True)
dfejecucion.sort_values(by=['Partida','Capitulo','Programa','Subtitulo','Item','Asignacion', 'Sub-Asignación'], inplace=True)
```

```python
dftemp = dfejecucion[((dfejecucion['Subtitulo']==24)&(dfejecucion['Item']==2))]
dfejecucion = dfejecucion.merge(dftemp, how='outer', indicator=True)
dfejecucion = dfejecucion[dfejecucion['_merge']=='left_only']
dfejecucion.drop('_merge', inplace = True, axis=1)

dftemp = dfejecucion[((dfejecucion['Subtitulo']==33)&(dfejecucion['Item']==2))]
dfejecucion = dfejecucion.merge(dftemp, how='outer', indicator=True)
dfejecucion = dfejecucion[dfejecucion['_merge']=='left_only']
dfejecucion.drop('_merge', inplace = True, axis=1)
```

```python
dfejec_acum = dfejecucion.groupby(by=['Ano','Mes','Moneda','Partida','Nombre Partida','Capitulo','Nombre Capitulo','Programa','Nombre Programa','Subtitulo','Nombre Subtitulo','Item','Nombre Item','Asignacion','Nombre Asignacion'])['Ejecución a AGOSTO'].sum()
dfejec_acum = dfejec_acum.reset_index()
```

```python
dfejec_acum['Ejecucion a Agosto en Pesos'] = dfejec_acum['Ejecución a AGOSTO']
dolares = [[2017,648],[2018,641],[2019,703],[2020,792],[2021,759],[2022, 859]]

for año, dolar in dolares:
    dfejec_acum.loc[(dfejec_acum['Ano'] == año)&(dfejec_acum['Moneda']=='DOLARES'), 'Ejecucion a Agosto en Pesos'] = dfejec_acum['Ejecución a AGOSTO'] * dolar

dfejec_acum
```

```python
csv_files = glob.glob(os.path.join(path + '/PV/', "*.csv"))
```

```python
dffinal = pd.DataFrame()
for f in csv_files:
    print(f)
    if f.rfind('PV2018.csv')>0:
        df = pd.read_csv(f, sep='\t', thousands=r'.')
        df.drop(columns=df.columns[-1:], axis=1,  inplace=True)

    elif f.rfind('PV2022.csv')>0:
        df = pd.read_csv(f, sep=',', thousands=r'.')
        df.drop('_id', axis=1, inplace=True)

    else:
        df = pd.read_csv(f, sep=';', thousands=r'.')
    
    df.columns = ['Ano','Mes','Moneda','Partida','Nombre Partida','Capitulo','Nombre Capitulo','Programa','Nombre Programa','Subtitulo','Nombre Subtitulo','Item','Nombre Item','Asignacion','Nombre Asignacion'\
        ,'Monto Ley Inicial','Monto Ley Vigente a AGOSTO']
    df['Item'].fillna(0, inplace=True)
    df['Asignacion'].fillna(0, inplace=True)
    df = df.sort_values(by=['Partida','Capitulo','Programa','Subtitulo','Item','Asignacion'])
    #df = df[(df['Partida']==6)&(df['Capitulo']==1)]
    #df = df[(df['Partida']==50)]#&(df['Subtitulo']==1)]#&(df['Programa']==1)]
    PCP = df[['Partida','Capitulo','Programa']].drop_duplicates()
    df_total = pd.DataFrame()
    for i, row in PCP.iterrows():
        df_temp = df[(df['Partida']==row['Partida'])&(df['Capitulo']==row['Capitulo'])&(df['Programa']==row['Programa'])]
        subtitulos = df_temp.Subtitulo.unique()
        df_prog = pd.DataFrame()
        monedas = df_temp.Moneda.unique()
        for moneda in monedas:
            df_mon = df_temp[df_temp['Moneda']==moneda]
            for subt in subtitulos:
                df_subt = df_mon[df_mon['Subtitulo']==subt]
                if len(df_subt) > 1:
                    df_subt_ceros = df_mon[df_mon['Subtitulo']==subt]
                    df_subt_ceros = df_subt_ceros[(df_subt_ceros['Asignacion']==0)&(df_subt_ceros['Item']==0)]
                    if len(df_subt_ceros) ==0:
                        df_subt_suma = df_subt.groupby('Subtitulo')['Monto Ley Vigente a AGOSTO'].sum().values[0]
                    else:
                        df_subt_suma = df_subt_ceros.groupby('Subtitulo')['Monto Ley Vigente a AGOSTO'].sum().values[0]

                    if len(df_subt[df_subt['Item']!=0]) == 0:
                        df_subt = df_subt.copy()
                    else:
                        df_subt = df_subt[df_subt['Item']!=0]
                    items = df_subt.Item.unique()
                    df_item_to = pd.DataFrame()
                    for item in items:
                        df_item = df_subt[df_subt['Item']==item]
                        #display(df_item)
                        #print(len(df_item))
                        if len(df_item) > 1:
                            if len(df_item[df_item['Asignacion']!=0]) == 0:
                                if len(df_item[df_item['Item']!=0])==0:
                                    df_item2 = df_item.copy()
                                else:
                                    df_item2 = df_item[df_item['Item']!=0]
                            else:
                                df_item2 = df_item[df_item['Asignacion']!=0]
                                try: 
                                    #print(df_item2.groupby('Item')['Monto Ley Vigente a AGOSTO'].sum().values[0], df_item.reset_index(drop=True)[:1]['Monto Ley Vigente a AGOSTO'].values[0])
                                    if df_item2.groupby('Item')['Monto Ley Vigente a AGOSTO'].sum().values[0] == df_item.reset_index(drop=True)[:1]['Monto Ley Vigente a AGOSTO'].values[0]:
                                        #print('1')
                                        df_item = df_item2.copy()
                                    else:
                                        #print(2)
                                        df_item.reset_index(drop=True, inplace=True)
                                        df_item.at[0, 'Asignacion'] = 9999
                                        df_item.at[0, 'Monto Ley Vigente a AGOSTO'] = df_item.reset_index(drop=True)[:1]['Monto Ley Vigente a AGOSTO'].values[0] - df_item2.groupby('Item')['Monto Ley Vigente a AGOSTO'].sum().values[0]
                                except:
                                    df_item = df_item[df_item['Asignacion']!=0]

                        #display(df_item)
                        
                        df_item_to = pd.concat([df_item, df_item_to], axis=0)

                    precalculo_subt = df_item_to.groupby('Subtitulo')['Monto Ley Vigente a AGOSTO'].sum().values[0]

                    if df_subt_suma != precalculo_subt:
                        df_subt_ceros.reset_index(drop=True, inplace=True)
                        df_subt_ceros.at[0, 'Asignacion'] = 9999
                        df_subt_ceros.at[0, 'Monto Ley Vigente a AGOSTO'] = df_subt_suma - precalculo_subt
                        df_item_to = pd.concat([df_subt_ceros, df_item_to], axis=0)
                        df_item_to.reset_index(drop=True, inplace=True)

                    df_prog = pd.concat([df_item_to, df_prog], axis=0)
                else:
                    df_prog = pd.concat([df_subt, df_prog], axis=0)

            df_total = pd.concat([df_prog, df_total], axis=0)

    df_total = df_total.sort_values(by=['Partida','Capitulo','Programa','Subtitulo','Item','Asignacion'])
    df_total.drop_duplicates(inplace=True)
    dffinal = pd.concat([df_total, dffinal], axis=0)
```

```python
dfpresu = dffinal.copy()
dfpresu['Ppto Vigente a Agosto en Pesos'] = dfpresu['Monto Ley Vigente a AGOSTO']
dfpresu = dfpresu.reset_index(drop=True)
for año, dolar in dolares:
    dfpresu.loc[(dfpresu['Ano'] == año)&(dfpresu['Moneda']=='DOLARES'), 'Ppto Vigente a Agosto en Pesos'] = dfpresu['Monto Ley Vigente a AGOSTO'] * dolar
```

```python
dftemp = dfpresu[((dfpresu['Subtitulo']==24)&(dfpresu['Item']==2))]
dfpresu = dfpresu.merge(dftemp, how='outer', indicator=True)
dfpresu = dfpresu[dfpresu['_merge']=='left_only']
dfpresu.drop('_merge', inplace = True, axis=1)

dftemp = dfpresu[((dfpresu['Subtitulo']==33)&(dfpresu['Item']==2))]
dfpresu = dfpresu.merge(dftemp, how='outer', indicator=True)
dfpresu = dfpresu[dfpresu['_merge']=='left_only']
dfpresu.drop('_merge', inplace = True, axis=1)
dftemp
```

```python
dfpresu_subt = dfpresu.groupby(['Ano','Partida','Nombre Partida','Capitulo','Nombre Capitulo','Programa','Nombre Programa','Subtitulo','Nombre Subtitulo'])['Ppto Vigente a Agosto en Pesos'].sum()
dfpresu_subt = dfpresu_subt.reset_index()

dfejec_subt = dfejec_acum.groupby(['Ano','Partida','Nombre Partida','Capitulo','Nombre Capitulo','Programa','Nombre Programa','Subtitulo','Nombre Subtitulo'])['Ejecucion a Agosto en Pesos'].sum()
dfejec_subt = dfejec_subt.reset_index()
```

```python
df_avance_ejec = dfpresu_subt.merge(dfejec_subt, how='outer', on=['Ano','Partida','Capitulo','Programa', 'Subtitulo'], indicator=True)
df_avance_ejec
```
